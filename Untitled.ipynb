```python
# Importing the Datasets 

df1 = pd.read_csv('Round1.csv')
df2 = pd.read_csv('Round2.csv')
df3 = pd.read_csv('Round3.csv')
df4 = pd.read_csv('Round4.csv')
df5 = pd.read_csv('Round5.csv')
df6 = pd.read_csv('Round6.csv')
#df7 = pd.read_csv('Round7.csv')
#df8 = pd.read_csv('met_nonmet_indicator.csv')
```

```python
# Merging the data from Round 1 - Round 4 with left merge (using original Round 1 population as reference for SPIDs)

mergedset1 = pd.merge(df1, df2, on='spid',how='left')
mergedset2 = pd.merge(mergedset1,df3,on='spid',how='left')
mergedset3 = pd.merge(mergedset2,df4,on='spid',how='left')

# Merging the data from R1-R4 with the data from Round 5 - Round 7 (with replenished population)
# Outer merge used to keep all entries in dataset from R1-R4 and add R5-R7 data without deleting any entries
# Data merged wherever spid match exists, but unmerged data also kept 

mergedset4 = pd.merge(mergedset3,df5,on='spid',how='outer')

mergedset5 = pd.merge(mergedset4,df6,on='spid',how='left')
mergedset_final = mergedset5

# Talk to Dr. Dasgupta about why merging Round 7 data results in deletion of certain columns?????
#mergedset6 = pd.merge(mergedset5,df7,on='spid',how='left')
#mergedset_final = pd.merge(mergedset6,df8,on='spid',how='left')
```

```python
propensity_set = hip_fracture_study[['spid','hc1brokebon1','r1d2intvrage','r1dgender','is1resptype','el1higstschl','rl1dracehisp','R1avg_income','r5dcontnew'#,'hc1disescn9','ss1prbbalcrd' 
                                    ]]
```

```python
test = propensity_set[propensity_set.hc1brokebon1 == 1]
control = propensity_set[propensity_set.hc1brokebon1 == 0]
test['hc1brokebon1'] = 1
control['hc1brokebon1'] = 0
```

```python
m = Matcher(test, control, yvar="hc1brokebon1", exclude=['spid'])
```

```python
m.fit_scores(balance=True, nmodels=150)
```

```python
m.predict_scores()
```

```python
m.plot_scores()
```

```python
m.tune_threshold(method='random')
```

```python
m.match(method="min", nmatches=1, threshold=0.0001)

m.record_frequency()
```
